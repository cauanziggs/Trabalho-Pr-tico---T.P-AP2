```python
class Processo:
    def __init__(self, nome):
        self.nome = nome
        self.relogio = 0

    def evento_interno(self):
        self.relogio += 1
        print(f"{self.nome}: Evento interno → Relógio = {self.relogio}")

    def enviar(self, destinatario):
        self.relogio += 1
        timestamp = self.relogio
        print(f"{self.nome}: Envia mensagem para {destinatario.nome} "
              f"(timestamp = {timestamp})")
        return timestamp

    def receber(self, remetente, timestamp_msg):
        self.relogio = max(self.relogio, timestamp_msg)
        self.relogio += 1
        print(f"{self.nome}: Recebe mensagem de {remetente.nome} "
              f"(timestamp recebido = {timestamp_msg}) → Relógio = {self.relogio}")


# Criar processos
P1 = Processo("P1")
P2 = Processo("P2")
P3 = Processo("P3")

print("\n Início da Simulação do Algoritmo de Lamport: \n")

# 1. P1 evento interno
P1.evento_interno()

# 2. P2 envia para P3
msg_2_3 = P2.enviar(P3)

# 3. P3 recebe de P2
P3.receber(P2, msg_2_3)

# 4. P1 envia para P2
msg_1_2 = P1.enviar(P2)

# 5. P3 evento interno
P3.evento_interno()

# 6. P2 recebe de P1
P2.receber(P1, msg_1_2)

# 7. P2 envia para P1
msg_2_1 = P2.enviar(P1)

# 8. P1 recebe de P2
P1.receber(P2, msg_2_1)

print("\n Fim da Simulação!!!")
```
